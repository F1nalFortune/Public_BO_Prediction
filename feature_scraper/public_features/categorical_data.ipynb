```python
def retry(ExceptionToCheck, tries=20, delay=3, backoff=2, logger=None):
  """
    Modified from source: http://wiki.python.org/moin/PythonDecoratorLibrary#Retry

    Objective
    ----------
    Exponential backoff function.

    Parameters
    ----------
    ExceptionToCheck : Exception or tuple
      the exception to check. may be a tuple of exceptions to check
      Possible values:
        ConnectionResetError, (TimeoutError, ConnectionError)
    tries : int
      number of times to try (not retry) before giving up
    delay : int
      initial delay between retries in seconds
    backoff: int
      backoff multiplier
      E.g. value of 2 will double the delay each retry
    logger : logging.Logger instance
      logger to use
  """
  def deco_retry(f):
    @wraps(f)
    def f_retry(*args, **kwargs):
      mtries, mdelay = tries, delay
      while mtries > 1:
        try:
          return f(*args, **kwargs)
        except ExceptionToCheck:
          msg = "%s, Retrying in %d seconds..." % (str(ExceptionToCheck), mdelay)
          if logger:
            #logger.exception(msg) # would print stack trace
            logger.warning(msg)
          else:
            print(msg)
          time.sleep(mdelay)
          mtries -= 1
          mdelay *= backoff
        return f(*args, **kwargs)
      return f_retry  # true decorator
  return deco_retry  
```

```python
import math
def genreData():
  #MACBOOK DATA
#     options = webdriver.ChromeOptions()
#     options.add_argument("no-sandbox")
#     options.add_argument("disable-dev-shm-usage")
#     options.add_argument("headless")
#     options.add_argument("user-data-dir=/Users/brianphelps/Library/Application Support/Google/Chrome/Thread_One")
#     driver = webdriver.Chrome(executable_path="/Users/brianphelps/Desktop/chromedriver", chrome_options=options)
  #BASEMENT OFFICE DATA
  options = webdriver.ChromeOptions()
  options.add_argument("no-sandbox")
  options.add_argument("disable-dev-shm-usage")
  # options.add_argument("headless")
  options.add_argument("user-data-dir=C:/Users/Brian/AppData/Local/Google/Chrome/User Data/X")
  driver = webdriver.Chrome("C:\\Users\Brian\Desktop\chromedriver.exe", chrome_options=options)


  @retry((ReadTimeout, ConnectTimeout, HTTPError, Timeout, ConnectionError), tries=20, delay=2, backoff=2)
  def serve_link():
    driver.get(f"https://www.boxofficemojo.com/genre/")
  serve_link()

  sourcez = driver.page_source
  html = BeautifulSoup(sourcez, 'lxml')

  genres = html.find('div', {"id": "table"}).find_all('table')[1].tbody.find_all('tr')
  genres = genres[1:]

  @retry((ReadTimeout, ConnectTimeout, HTTPError, Timeout, ConnectionError), tries=20, delay=2, backoff=2)
  def movie_link(link):
    driver.get(link)

  genre_df = pd.DataFrame()
  for genre in genres:
    link = genre.td.a['href']
    genre_name = genre.td.a.text.strip()
    total_titles = int(genre.find_all('td')[2].text.replace(",", "").strip())
    clicks = math.ceil(int(total_titles)/100)
    
    movie_link(f"https://www.boxofficemojo.com/{link}")

    genre_source = driver.page_source
    genre_html = BeautifulSoup(genre_source, 'lxml')

    movies = genre_html.find('div', {"id": "table"}).find_all('table')[1].tbody.find_all('tr')
    movies= movies[1:]
    if clicks > 1:
      for i in range(clicks):
        if(i+1)<clicks:
          click_button = genre_html.find(text="Next page").parent['href']
          next_link = f"https://www.boxofficemojo.com/{click_button}"      
        for movie in movies:
          values = movie.find_all('td')
          try:
            if values[1].div.span:
              continue
          except Exception as e:
              name = values[1].a.text.strip()
              date = values[6].text
              profile = values[1].a['href'].split("?ref")[0]
              temp = pd.DataFrame({
                'name': [name],
                'date': [date],
                'genre': [genre_name],
                'profile': [profile]
              })
              genre_df = genre_df.append(temp)
        if(i+1)<clicks:
          movie_link(next_link)
          genre_source = driver.page_source
          genre_html = BeautifulSoup(genre_source, 'lxml')
          time.sleep(3)
    else:
      for movie in movies:
        values = movie.find_all('td')
        try:
          if values[1].div.span:
            continue
        except Exception as e:
            name = values[1].a.text.strip()
            date = values[6].text
            profile = values[1].a['href'].split("?ref")[0]
            temp = pd.DataFrame({
              'name': [name],
              'date': [date],
              'genre': [genre_name],
              'profile': [profile]
            })
            genre_df = genre_df.append(temp)
  driver.quit()
  genre_df = genre_df.reset_index()
  return genre_df
```
